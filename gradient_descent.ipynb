```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_pred(x,w,b):
    return w*x + b
```

```python
def compute_gradient(x,y,y_pred):
    m = len(x)
    d_w,d_b = 0,0
    for i in range(m):
        d_w = (1/m)*np.sum(x*(y_pred - y)**2)
        d_b = (1/m)*np.sum((y_pred - y)**2)
    return d_w,d_b
```

```python
def gradient_descent(x,y,w,b,r,it):
    for i in range(it):
        y_pred = compute_pred(x,y,b)
        d_w,d_b = compute_gradient(x,y,y_pred)
        w -= r*d_w
        b -= r*d_b
    return w,b
```

```python
x = np.array([1.0, 2.0])
y = np.array([300.0, 500.0])
w,b = 0,0
r = -0.00001
it = 66
w,b = gradient_descent(x,y,w,b,r,it)
print(w,b)
```

```python
plt.scatter(x,y)
y_pred = compute_pred(x,w,b)
plt.plot(x,y_pred)
```
